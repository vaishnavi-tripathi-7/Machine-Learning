```python
data = pd.read_csv('barchart_data.csv')
index = data['Responder_id']
response = data['LanguagesWorkedWith']
count = Counter()

for row in response :
    count.update(row.split(';'))

top_10 = count.most_common(10)
top_10

lang = []
val = []
for i in top_10 :
    lang.append(i[0])
    val.append(i[1])

plt.barh( lang , val)
plt.xlabel("Number of Respondents")
plt.title("Top 10 Programming Languages Worked With")
plt.show()
```

# Stack Plots

```python
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [0, 1, 1, 2, 2, 3, 3, 3, 4]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels = ['player1', 'player2', 'player3']
colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.stackplot(minutes , player1 , player2 , player3 , labels = labels, colors = colors)
plt.legend(loc = 'upper left')

plt.tight_layout()
plt.show()
```

```python
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

dev1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
dev2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
dev3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels = ['dev1', 'dev2', 'devr3']
colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.stackplot(minutes , dev1 , dev2 , dev3 , labels = labels, colors = colors)
plt.legend(loc = 'lower left')

plt.tight_layout()
plt.show()
```

# Timeseries

```python
dates = [datetime(2019, 5, 24), datetime(2019, 5, 25), datetime(2019, 5, 26), datetime(2019, 5, 27), datetime(2019, 5, 28), datetime(2019, 5, 29), datetime(2019, 5, 30)]
y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates,y , linestyle = 'solid' , linewidth=0.5 , marker='X')

plt.gcf().autofmt_xdate()      # for date alignment on x axis

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
```

```python
data = pd.read_csv('timeseries_matplotlib.csv')
price_date = data['Date']          # data stored as string
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid' , linewidth=0.5 , marker='X')
plt.gcf().autofmt_xdate()   
plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
```

```python
data = pd.read_csv('subplot_matplotlib.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, py_salaries, label='Python' , linewidth = 0.5)
plt.plot(ages, js_salaries, label='JavaScript' , linewidth = 0.5)
plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs' , linewidth = 0.5)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
```
